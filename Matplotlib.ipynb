```python
#A scatter plot is a type of data visualization that uses Cartesian coordinates 
#to display values for typically two variables for a set of data. 
#The data is represented by points (or markers) on a two-dimensional plane, where the position 
#of each point is determined by the values of the two variables. 
#Scatter plots are useful for identifying patterns, correlations, 
#and trends in data, as well as for detecting outliers.
```

```python
# The subplot() function in Matplotlib is used to create multiple plots in a single figure, 
# allowing for the comparison of different datasets or different views of the same 
# dataset within one graphical window. This function is particularly useful for creating c
# omplex visualizations that require the display of multiple plots in an organized 
# layout, such as a grid of plots.
```

```python
x = np.array([0, 1, 2, 3, 4, 5])  
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.figure(figsize=(10, 8))

# First subplot
plt.subplot(2, 2, 1) # (rows, columns, index)
plt.plot(x, y1)
plt.title('Plot of line 1')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

# Second subplot
plt.subplot(2, 2, 2) # (rows, columns, index)
plt.plot(x, y2)
plt.title('Plot of line 2')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

# Third subplot
plt.subplot(2, 2, 3) # (rows, columns, index)
plt.plot(x, y3)
plt.title('Plot of line 3')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

# Fourth subplot
plt.subplot(2, 2, 4) # (rows, columns, index)
plt.plot(x, y4 )
plt.title('Plot of line4')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

# Adjust layout to prevent overlap
plt.tight_layout()

# Show plots
plt.show()




```

```python
# A bar plot (or bar chart) is a graphical representation of data that uses rectangular bars 
# to show the frequency, count, or other measures (e.g., sum, average) 
# for different categories. The length or height of each bar is proportional 
# to the value it represents. Bar plots can be plotted vertically or horizontally.
```

```python
# Comparison: Bar plots are particularly useful for comparing the values of different categories. Each bar represents a category, making it easy to compare their values visually.
# Categorical Data: They are ideal for displaying and comparing categorical data. Each bar represents a category, and its height (or length) represents the value associated with that category.
# Distribution: Bar plots can show the distribution of a categorical variable, making it easy to see which categories are most or least common.
# Easy Interpretation: They are straightforward to interpret, making them accessible to a wide audience, including those without a statistical background.
# Versatility: Bar plots can be used in various fields, such as business, science, and social sciences, to present data in a clear and concise manner.
```

```python
# A box plot, also known as a box-and-whisker plot, 
# is a standardized way of displaying the distribution of data based on a 
# five-number summary: minimum, first quartile (Q1), median, third quartile (Q3), and maximum. Box plots are useful for summarizing a large dataset and identifying patterns, including central tendency, dispersion, and outliers.

# Key Components of a Box Plot:
# Box: The main part of the plot, representing the interquartile range (IQR), which contains the middle 50% of the data. The bottom and top edges of the box indicate the first and third quartiles (Q1 and Q3), respectively.
# Median Line: A line inside the box representing the median (the middle value of the dataset).
# Whiskers: Lines extending from the box to the minimum and maximum values within 1.5 times the IQR from the quartiles. They help to show the spread of the data outside the middle 50%.
# Outliers: Individual points plotted outside the whiskers, representing data points that fall outside the typical range (beyond 1.5 times the IQR from Q1 or Q3).


# Why is a Box Plot Used?
# Data Distribution: Box plots provide a clear visual summary of the distribution of a dataset, highlighting its central value, spread, and symmetry.
# Comparison: They are effective for comparing distributions across different categories or groups. Multiple box plots can be plotted side-by-side for comparison.
# Outlier Detection: Box plots help in identifying outliers, which are data points that deviate significantly from the rest of the data.
# Summary Statistics: They succinctly display key summary statistics of a dataset, providing insight into its central tendency and variability without making any assumptions about the underlying statistical distribution.
# Data Skewness: Box plots indicate skewness in the data by showing the relative positions of the median and the quartiles.
```

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Create box plot
plt.boxplot([box1, box2], vert=True, patch_artist=True, labels=['Box 1', 'Box 2'])

# Add title and labels
plt.title('Box Plot for Two Datasets')
plt.xlabel('Dataset')
plt.ylabel('Values')

# Show plot
plt.show()
```
